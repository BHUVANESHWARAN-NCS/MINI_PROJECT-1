```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\adminR\orders.csv")
df
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df = df.rename(columns = {'Order Id':'order_id',
             'Order Date':'order_date',
             'Ship Mode':'ship_mode',
             'Segment':'segment',
             'Country':'country',
'City':'city',
'State':'state',
'Postal Code':'postal_code',
'Region	Category':'region_category',
'Sub Category':'sub_category',
'Product Id':'product_id',
'cost price':'cost_price',
'List Price':'list_price',
'Quantity':'quantity',
'Discount Percent':'discount_percent'})
```

```python
df.columns.str.strip()
print(df.columns)
```

```python
df['discount_price'] = df['list_price'] * df['discount_percent'] / 100
df['sales_price'] = df['list_price'] - df['discount_price']
df['profit'] = df['sales_price'] - df['cost_price']
```

```python
print(df.columns)
```

```python
df=df.to_csv(r"C:\Users\adminR\retail_order.csv")
```
